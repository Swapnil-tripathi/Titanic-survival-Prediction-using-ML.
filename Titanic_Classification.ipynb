**Introduction :**





▪The RMS Titanic was one of the most famous ship in the maritime history.It was esteemed for its sheer size,opulence and maiden voyage that ended in disaster

▪The Titanic was built  at the Harland and Wolf shipyard in Belfast,Northern Ireland.

**About the Dataset :**

The dataset contains information about 891 passengers traveeling in the ship.
Following are the important features of the dataset:





1.   PassengerId : Id of passenger
2. Survived : This column indicates whether a passenger survived or not
            1 denotes survival of the passenger whereas 0 denotes passenger didn't survived.
3. Pclass : Its full form is passenger class and it denotes the class
          of passenger's ticket(1st,2nd or 3rd).
4. Name : The name of the passenger.
5. Sex : Gender of the passenger(Male or Female).
6. Age : Age of the passenger.
7. SibSp : Number of siblings or spouses the passenger had on board.
8. Parch : Number of parent or children the passenger had on board.
9. Ticket : Ticket number of the passenger.
10. Fare : Amount paid by the passenger for the ticket.
11. Cabin : Cabin number where the passenger stayed.
12. Embarked : The port at which passenger boarded the Titanic(C=Cherbourg,Q=Queenstown,S=Southampton)

**Label Encoding:**

Label Encoding is a technique used to convert these categorical variables into numerical labels.

**Features and Target variable**

```python
cols=['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']
features=titanic[cols]
target=titanic['Survived']
```

```python
#Importing libraries
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
#Train_Test_Split
features_train,features_test,target_train,target_test=train_test_split(features,target,test_size=0.3,random_state=5)
```

```python
print(features_train.shape)
print(features_test.shape)
print(target_train.shape)
print(target_test.shape)
```

Defining a function to compute Confusion Matrix,Classification report
and to generate training and the testing score( to calculate Accuracy)

```python
#Importing important libraries
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
def cls_eval(targettest,targetpred):
  cm=confusion_matrix(targettest,targetpred)
  print('Confusion Matrix\n',cm)
  print('Classification Report\n',classification_report(targettest,targetpred))


def mscore(model):
  print('Training_score',model.score(features_train,target_train))#Training accuracy
  print('Testing_score',model.score(features_test,target_test))#Testing accuracy
```

## ***Logistic Regression***

```python
model=LogisticRegression()#Selecting the model
model.fit(features_train,target_train)
mscore(model)#Computing Training and Testing score
```

```python
#Generating Prediction
prediction=model.predict(features_test)
print(prediction)
```

```python
#Evaluation of the model : Printing Confusion matrix,Classification report and Accuracy score.
cls_eval(target_test,prediction)
accuracy_lr=accuracy_score(target_test,prediction)
print('Accuracy Score : ',accuracy_lr)
```

## ***KNN Classifier Model***

```python
model=KNeighborsClassifier()
model.fit(features_train,target_train)
mscore(model)#Computing Training Score and Testing Score
```

```python
#Generating predictions
prediction_knn=model.predict(features_test)
print(prediction_knn)
```

```python
#Evaluating the model by using Confusion matrix,Classification report and accuracy score
cls_eval(target_test,prediction_knn)
accuracy_knn=accuracy_score(target_test,prediction_knn)
print('Accuracy Score : ',accuracy_knn)
```

## ***Support Vector Classifier Model***

```python
model=SVC(C=1.0)
model.fit(features_train,target_train)
mscore(model)
```

```python
#Generating Prediction
prediction_svc=model.predict(features_test)
print(prediction_svc)
```

```python
#Evaluation of the model with the help of Confusion matrix,Classification report and Accuracy score
cls_eval(target_test,prediction_svc)
accuracy_svc=accuracy_score(target_test,prediction_svc)
print('Accuracy score : ',accuracy_svc)
```

## ***Random Forest Classifier Model***

```python
model=RandomForestClassifier(n_estimators=80,criterion='entropy',min_samples_split=5,max_depth=10)#Selecting the model
model.fit(features_train,target_train)
mscore(model)#Calculating Training and Testing score
```

```python
#Generating Predictions
prediction_randomforest=model.predict(features_test)
print(prediction_randomforest)
```

```python
#Evaluating the model with the help of Confusion matrix,Classification report and Accuracy score
cls_eval(target_test,prediction_randomforest)
accuracy_rf=accuracy_score(target_test,prediction_randomforest)
print('Accuracy Score : ',accuracy_rf)
```

##  ***Decision Tree Classifier Model***

```python
model=DecisionTreeClassifier(max_depth=5,criterion='entropy',min_samples_split=10)#Selecting the model
model.fit(features_train,target_train)
mscore(model)#Calculating Training and Testing Score
```

```python
#Generating Predictions
prediction_decisiontree=model.predict(features_test)
print(prediction_decisiontree)
```

```python
#Evaluating the model with the help of Confusion matrix,Classification report and accuracy score
cls_eval(target_test,prediction_decisiontree)
accuracy_dt=accuracy_score(target_test,prediction_decisiontree)
print('Accuracy Score : ',accuracy_dt)
```

## ***Adaboost Classifier Model***

```python
model=AdaBoostClassifier(n_estimators=80)#Selecting the model
model.fit(features_train,target_train)
mscore(model)#Calculating Training and Testing score
```

```python
#Predictions
prediction_adaboost=model.predict(features_test)
print(prediction_adaboost)
```

```python
#Confusion matrix,Classification report and Accuracy score
cls_eval(target_test,prediction_adaboost)
accuracy_adaboost=accuracy_score(target_test,prediction_adaboost)
print('Accuracy score : ',accuracy_adaboost)
```

## ***Comparing Accuracy score of all the models used in this project***

```python
models=pd.DataFrame({
    'Model': ['Logistic Regression','knn','SVC','Random Forest Classifier','Decision Tree Classifier','Ada Boost Classifier'],
    'Score': [accuracy_lr,accuracy_knn,accuracy_svc,accuracy_rf,accuracy_dt,accuracy_adaboost]
})
models.sort_values(by='Score',ascending=False)
```

We can see that Adaboost classifier has the best accuracy score among all the models used in it.
